```python
df_country=pd.read_excel('country-code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# pie chart - top 3 country uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:Zomato maximum records or transactions are form india after that is from usa and then Uk(united kingdom)

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

## Observations
1. when rating is between 4.5 to 4.9 ------> exellent
2. when ratings are betweem 4.0 to 4.4 -----> very good
3. when ratings are betweem 3.5 to 3.9   -----> good
4. when ratings are betweem 2.5 to 3.4 -----> average
5. when ratings are betweem 1.8  to 2.4   -----> poor

```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color', data=ratings,palette=['Red','orange','yellow','green'])
```

# Observations
* 1. Not Rated count is very high 
* 2. Maximum number of a rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x='Rating color',data=ratings,palette=['Red','orange','yellow','green'])
```

```python
## find out which currency is used by which country 
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
# Find out which country have online delivery or not ?
final_df[final_df['Has Online delivery']=="Yes"].Country.value_counts()
```

```python
final_df[['Country','Has Online delivery']] \
    .groupby(['Country','Has Online delivery']) \
    .size() \
    .reset_index(name='Count') \
    .sort_values(['Country','Has Online delivery'])
```

# Observations :-
* 1. Online delivery is only available in india and UAE
* 2. And some part of india online delivery is not available
